```python
import pandas as pd
import numpy as np
```

```python
# read csv 
df = pd.read_csv(r'C:\Users\WeilunShi\Dropbox (IFPRI)\CML-Ethiopia - team folder\Operation\combined_rain_df.csv')
```

```python
# filter out df with c1 has value
df = df[df['c1'].notnull()]

# add an indicator if c1_a or c1_b is not null
df['c1_a_ind'] = np.where(df['c1_a'].notnull(), 1, 0)
df['c1_b_ind'] = np.where(df['c1_b'].notnull(), 1, 0)
df['has_value'] = np.where((df['c1_a_ind'] == 1) | (df['c1_b_ind'] == 1), 1, 0)

# group by date and respondent_id, count the number of observations
newdf=df.groupby(['date', 'respondent_id'])['has_value'].sum().reset_index(name='day_rain_count')

# merge the newdf with the original df
df = pd.merge(df, newdf, on=['date', 'respondent_id'], how='left')
```

```python
# filter out day_rain_count != c1
df_inconsistent = df[df['day_rain_count'] != df['c1']]
```

```python
# drop c1_a_ind, c1_b_ind, has_value
df_inconsistent = df_inconsistent.drop(['c1_a_ind', 'c1_b_ind', 'has_value'], axis=1)
```

```python
# rename column index as row_id
df_inconsistent = df_inconsistent.rename(columns={'index': 'row_id'})
```
