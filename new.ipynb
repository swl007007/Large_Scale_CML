```python
# select unique Pixel_Longitude_c and Pixel_Latitude_c from df2
location = df2[['Pixel_Longitude_c', 'Pixel_Latitude_c']].drop_duplicates()
```

```python
# rename Pixel_Longitude_c and Pixel_Latitude_c to lon and lat
location = location.rename(columns={'Pixel_Longitude_c': 'lon', 'Pixel_Latitude_c': 'lat'})
```

```python
adminBoundaries = ee.FeatureCollection("FAO/GAUL/2015/level2").filter(ee.Filter.eq('ADM1_NAME', 'Amhara'))
# use lat and lon to filter adminBoundaries
adminBoundaries_1 = adminBoundaries.filterBounds(ee.Geometry.Point(location.iloc[0]['lon'], location.iloc[0]['lat']))

# initialize our map
map2 = geemap.Map(center=[location.iloc[0]['lat'],location.iloc[0]['lon']], zoom=9)
map2.add_basemap('SATELLITE')

map2.addLayer(adminBoundaries_1, {}, "Covered_Boundaries", opacity=0.75)

map2.addLayerControl()
map2
# convert adminBoundaries_1 to geojson
adminBoundaries_1_geojson = adminBoundaries_1.getInfo()
```

```python
# plot the map
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, Polygon
import contextily as ctx
boundaries = gpd.read_file(r"C:\Users\WeilunShi\Dropbox (IFPRI)\CML-Ethiopia - team folder\CML_50_Towns_Results\eth_adm_csa_bofedb_2021_shp\eth_admbndl_admALL_csa_bofedb_itos_2021.shp")
# create a geometry column from coordinates

# Coordinate reference system : WGS84
crs = {'init': 'epsg:4326'}

# create GeoDataFrame for adminBoundaries_1_geojson
geo_df = gpd.GeoDataFrame.from_features(adminBoundaries_1_geojson['features'], crs=crs)

# plot the map
fig, ax = plt.subplots(figsize = (15,15))
boundaries.plot(ax=ax, alpha=0, color='grey')
geo_df.plot(ax=ax, markersize=15, color='red', marker='x', label='grid')
plt.legend(prop={'size':15})
ctx.add_basemap(ax, crs=geo_df.crs.to_string(), source=ctx.providers.Stamen.TonerLite)
plt.title('CML Locations in Ethiopia', fontsize=20)
plt.xlabel('Longitude', fontsize=15)
plt.ylabel('Latitude', fontsize=15)

# zoom in
ax.set_xlim(32.5, 45)
ax.set_ylim(3, 15)




plt.show()
```
